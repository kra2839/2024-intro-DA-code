```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig5=plt.figure(concatenated_layout=True)
widths=[2,3,1.5]
heights=[1,3,2]

spec5=fig5.add_gridspec(ncols=3,nrows=3,width_ratios=widths,height_ratios=heights)
for row in range(3):
    for col in range(3):
        ax=fig5.add_subplot(spec5[row,col])
        label= 'width:{}/height:{}'.format(widths[col],heights[row])
        ax.annotate(label,(0.1,0.5),xycoords='axes fraction',va='center',ha='center')
```
